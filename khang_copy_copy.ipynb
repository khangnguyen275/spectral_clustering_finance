```python
from utils.cluster import *

def windsorize_corr_matrix(corr, lower_pct=0.01, upper_pct=0.99):
    """
    Windsorize the correlation matrix by capping values below the lower percentile
    and above the upper percentile.
    
    Parameters:
        corr (np.ndarray or pd.DataFrame): Correlation matrix.
        lower_pct (float): Lower percentile (between 0 and 1).
        upper_pct (float): Upper percentile (between 0 and 1).
        
    Returns:
        np.ndarray: Windsorized correlation matrix.
    """
    corr_values = corr.values.flatten() if hasattr(corr, 'values') else corr.flatten()
    lower = np.percentile(corr_values, lower_pct * 100)
    upper = np.percentile(corr_values, upper_pct * 100)
    corr_windsorized = np.clip(corr, lower, upper)
    return corr_windsorized
```

```python
def calculate_cut(W, C, cluster_idx):
    """
    Calculate the cut value for a given cluster.
    W: Adjacency matrix (numpy array)
    C: Cluster indicator vector
    cluster_idx: Index of the cluster to calculate the cut for
    """
    mask_in = (C == cluster_idx)
    mask_out = ~mask_in
    cut_value = np.sum(W[mask_in][:, mask_out])
    return cut_value


def calculate_vol(W, C, cluster_idx):
    """
    Calculate the volume for a given cluster.
    W: Adjacency matrix (numpy array)
    C: Cluster indicator vector
    cluster_idx: Index of the cluster to calculate the volume for
    """
    mask_in = (C == cluster_idx)
    vol_value = np.sum(W[mask_in])
    return vol_value


def object_function(W, C, tau_plus, tau_minus):
    """
    Calculate the minimizing function based on the given parameters.
    W: Adjacency matrix (numpy array)
    C: Cluster indicator vector (numpy array)
    tau_plus, tau_minus: Trade-off parameters
    """
    clusters = np.unique(C)
    total = 0
    G_plus = np.maximum(W, 0)
    G_minus = np.maximum(-W, 0)
    for cluster_idx in clusters:
        cut_plus = calculate_cut(G_plus, C, cluster_idx)
        cut_minus = calculate_cut(G_minus, C, cluster_idx)
        vol_plus = calculate_vol(G_plus, C, cluster_idx)
        vol_minus = calculate_vol(G_minus, C, cluster_idx)
        term = (cut_plus + tau_minus * vol_minus) / (cut_minus + tau_plus * vol_plus)
        total += term
    #     print(f"Cluster {cluster_idx}: Term = {term}")
    # print(f"Object function value: {total}")
    return total

def get_cluster_sizes_list(cluster_vector):
    """
    Returns a list of cluster sizes, in the order of sorted unique cluster labels.
    
    Parameters:
        cluster_vector (array-like): Cluster assignments for each item.
        
    Returns:
        list: List of cluster sizes.
    """
    unique, counts = np.unique(cluster_vector, return_counts=True)
    return list(counts)

def sum_edges_within_cluster(W, C, cluster_idx):
    """
    Calculate the sum of edge weights within a given cluster.
    
    Parameters:
        W (np.ndarray): Adjacency (correlation) matrix.
        C (array-like): Cluster assignment vector.
        cluster_idx (int): The cluster label to compute the sum for.
        
    Returns:
        float: Sum of edge weights within the cluster.
    """
    mask = (C == cluster_idx)
    # Select the submatrix for the cluster and sum all elements
    return np.sum(W[mask][:, mask])

def correlation_object_function(W, C):
    """
    Calculate the minimizing function based on the given parameters.
    W: Adjacency matrix (numpy array)
    C: Cluster indicator vector (numpy array)
    tau_plus, tau_minus: Trade-off parameters
    """
    clusters = np.unique(C)
    total = 0
    cluster_size_list = get_cluster_sizes_list(C)
    for cluster_idx in clusters:
        cluster_size = cluster_size_list[cluster_idx]
        term = (sum_edges_within_cluster(W, C, cluster_idx) - cluster_size) / (cluster_size ** 2 - cluster_size)
        print(cluster_size_list[cluster_idx])
        total += term
        print(f"Cluster {cluster_idx}: Term = {term}")
    print(f"Object function value: {total}")
    return total

def correlation_object_list(W, C):
    """
    Calculate the minimizing function based on the given parameters.
    W: Adjacency matrix (numpy array)
    C: Cluster indicator vector (numpy array)
    tau_plus, tau_minus: Trade-off parameters
    """
    clusters = np.unique(C)
    scores = []
    cluster_size_list = get_cluster_sizes_list(C)
    for cluster_idx in clusters:
        cluster_size = cluster_size_list[cluster_idx]
        term = (sum_edges_within_cluster(W, C, cluster_idx) - cluster_size) / (cluster_size ** 2 - cluster_size)
        scores.append(term)
        print(f"Cluster {cluster_idx}: Term = {term}")
    return scores
```

## Modify the Trading Algorithm

```python
def identify_stocks(R_curr: pd.DataFrame):
    R_curr = R_curr.copy()

    # On each day, calculate the deviation from the mean of the cluster for each stock
    numeric_cols = R_curr.columns.difference(['ticker', 'cluster'])
    R_curr[numeric_cols] = (
        R_curr
        .groupby('cluster')[numeric_cols]
        .transform(lambda col: col - col.mean())
    )

    # Drop the non-numeric columns, then sum all deviations within the sliding window
    R_curr['deviation'] = R_curr.drop(columns=['ticker', 'cluster']).sum(axis=1)

    # Identify winners and losers based on the threshold value p
    threshold = 0

    # Start with zeros
    R_curr['trade'] = 0

    # Set +1 where deviation exceeds threshold signifying winners
    R_curr.loc[R_curr['deviation'] > threshold, 'trade'] = 1

    # Set -1 where deviation is below threshold signifying losers
    R_curr.loc[R_curr['deviation'] < threshold, 'trade'] = -1

    return R_curr

def assign_stock_weights(R_curr: pd.DataFrame):

  R_curr = R_curr.copy()

  # Calculate nK and mK for each cluster
  cluster_counts = R_curr.groupby('cluster')['trade'].value_counts().unstack(fill_value=0)

  # Calculate the notional values
  # If trade is +1, notional is 1 / number of +1 trades in the cluster
  # If trade is -1, notional is 1 / number of -1 trades in the cluster
  # If trade is 0, notional is 0
  R_curr['notional'] = 0.0  # Initialize the 'notional' column with zeros

  for cluster_id in cluster_counts.index:
      nK = cluster_counts.loc[cluster_id].get(1, 0)  # Get count of +1 trades, default to 0 if no +1
      mK = cluster_counts.loc[cluster_id].get(-1, 0) # Get count of -1 trades, default to 0 if no -1

      # Assign notional for +1 trades in the current cluster, handling division by zero
      if nK > 0:
          R_curr.loc[(R_curr['cluster'] == cluster_id) & (R_curr['trade'] == 1), 'notional'] = -1

      # Assign notional for -1 trades in the current cluster, handling division by zero
      if mK > 0:
          R_curr.loc[(R_curr['cluster'] == cluster_id) & (R_curr['trade'] == -1), 'notional'] = 1 

  # Display the updated DataFrame
  return R_curr

def execute_trading_strategy(win_threshold: float,
                            lookback_window = 60,
                            lookforward_window = 3,
                            w = 5,
                            eligible_dates = None,
                            cl_med = 'SPONGE',
                            num_med = 'var'):
    # record the total number of days
    num_dates = len(eligible_dates)
    # first trading day
    current_date = lookback_window
    # record daily_PnL
    daily_PnL = []

# renew_portfolio criteria
    # trading_period = 0
    # trading_PnL = 0
    # update_portfolio = True

    while current_date + lookforward_window < 200:
        start_date = current_date - lookback_window
        # size of R_curr: #stocks x (1 ticker + 63 days)
        # size of market_curr: 63
        R_curr, market_curr = get_sliding_window_data(eligible_dates = eligible_dates,
                                                    lookback_window = lookback_window,
                                                    lookforward_window=lookforward_window,
                                                    start_date = start_date)
        # R_cov is the matrix containing the return in the 60 days lookback window
        R_cov = R_curr.iloc[:, : lookback_window + 1]
        market_cov = market_curr[: lookback_window]
        # clusterize the stocks
        R_cov = clusterize(cl_med, num_med, R_cov, market_cov)
        R_cov = assign_stock_weights(identify_stocks(R_cov))
        # calculate PnLs for the lookforward window
        bet_size = R_cov['notional'].to_numpy()
        lookback = -1 * lookforward_window
        future_return = R_curr.iloc[:, lookback:].to_numpy()
        PnLs = future_return.T @ bet_size
        Cumpnl = np.cumsum(PnLs)

        if np.max(Cumpnl) > win_threshold:
            index = np.argmax(Cumpnl > win_threshold)
            PnLs = PnLs[:index+1]

            daily_PnL += PnLs.tolist()
            current_date += index + 1
        else:
            daily_PnL += PnLs.tolist()
            current_date += lookforward_window
        print(f"Day {current_date+1}: PnL = {PnLs}")
    return daily_PnL
```

```python
def select_top_half_clusters(corr_matrix, cluster_vector):
    """
    Selects the top half of clusters with the lowest correlation_object_function term values.
    Returns a mask (boolean array) indicating which stocks belong to the selected clusters.
    """
    unique_clusters = np.unique(cluster_vector)
    # Compute the term for each cluster
    cluster_terms = []
    for cluster_idx in unique_clusters:
        mask = (cluster_vector == cluster_idx)
        cluster_size = np.sum(mask)
        if cluster_size > 1:
            term = (np.sum(corr_matrix[mask][:, mask]) - cluster_size) / (cluster_size ** 2 - cluster_size)
        else:
            term = np.nan  # Avoid division by zero for singleton clusters
        cluster_terms.append((cluster_idx, term))
    # Sort clusters by term (ascending, lower is better)
    sorted_clusters = sorted([ct for ct in cluster_terms if not np.isnan(ct[1])], key=lambda x: x[1])
    # Select top half
    n_select = len(sorted_clusters) // 2
    selected_clusters = set([cl for cl, _ in sorted_clusters[:n_select]])
    # Create mask for stocks in selected clusters
    selected_mask = np.array([cl in selected_clusters for cl in cluster_vector])
    return selected_mask, selected_clusters
```

```python
def select_top_half_clusters_largest(corr_matrix, cluster_vector):
    """
    Selects the top half of clusters with the LARGEST correlation_object_function term values.
    Returns a mask (boolean array) indicating which stocks belong to the selected clusters.
    """
    unique_clusters = np.unique(cluster_vector)
    # Compute the term for each cluster
    cluster_terms = []
    for cluster_idx in unique_clusters:
        mask = (cluster_vector == cluster_idx)
        cluster_size = np.sum(mask)
        if cluster_size > 1:
            term = (np.sum(corr_matrix[mask][:, mask]) - cluster_size) / (cluster_size ** 2 - cluster_size)
        else:
            term = np.nan  # Avoid division by zero for singleton clusters
        cluster_terms.append((cluster_idx, term))
    # Sort clusters by term (descending, larger is better)
    sorted_clusters = sorted([ct for ct in cluster_terms if not np.isnan(ct[1])], key=lambda x: x[1], reverse=True)
    # Select top half
    n_select = len(sorted_clusters) // 2
    selected_clusters = set([cl for cl, _ in sorted_clusters[:n_select]])
    # Create mask for stocks in selected clusters
    selected_mask = np.array([cl in selected_clusters for cl in cluster_vector])
    return selected_mask, selected_clusters
```
